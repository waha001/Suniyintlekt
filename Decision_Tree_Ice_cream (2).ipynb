```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
shox= {
    'Harorat (°C)': [20, 25, 30, 35, 40],
    'narxi ($)': [100, 150, 200, 250, 300]
}

# Convert the data into a pandas DataFrame
df = pd.DataFrame(shox)
```

```python
df
```

```python
# Features and Target
X = df[['Harorat (°C)']]  # Feature (input data)
y = df['narxi($)']  # Target variable (output data)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.tree import plot_tree
```

```python
# Train the Decision Tree Regressor
tree= DecisionTreeRegressor()  # Limit depth to avoid overfitting
tree.fit(X, y)
```

```python
# New data for prediction (new temperature)
new_temperature = pd.DataFrame({'Harorat (°C)': [26]})
```

```python
# Predict the ice cream sales for the new temperature
predicted_sales = tree.predict(new_temperature)
```

```python
# Output the predicted ice cream sales
print("Predicted Ice Cream Sales for 33°C:", predicted_sales[0])
```

```python
X.columns
```

```python
df
```

```python
# Visualize the Decision Tree
plt.figure(figsize=(8, 6))  # Set the figure size for better readability
plot_tree(tree, feature_names=X.columns, filled=True, rounded=True)
plt.title("Decision Tree for Ice Cream Sales Prediction")
plt.show()
```

```python
# Create a DataFrame (dataset) with house sizes, number of rooms, and prices
shox= {
    'Kvadrat': [65, 70, 80, 86, 100, 45],
    'Hona': [2, 3, 3, 4, 4, 1],
    'Narx': [15000, 19000, 25000, 45000, 60000, 12000]
}

# Convert the data into a pandas DataFrame
df = pd.DataFrame(shox)
```

```python
df
```

```python
# Features and Target
X = df[['Kvadrat', 'Hona']]  # Features (input data)
y = df['Narx']  # Target variable (output data)
```

```python
# Train the Decision Tree Regressor
tree = DecisionTreeRegressor(max_depth=2)  # Limit depth to avoid overfitting
tree.fit(X, y)
```

```python
# New data for prediction (new house with size and number of rooms)
new_house = pd.DataFrame({
    'Kvadrat': [82],
    'Hona': [3]
})
```

```python
# Predict the price of the new house
predicted_price = tree.predict(new_house)
```

```python
# Output the predicted price
print("Predicted Price for the new house:", predicted_price[0])
```

```python
# Visualize the Decision Tree
plt.figure(figsize=(12, 8))  # Set the figure size for better readability
plot_tree(tree, feature_names=X.columns, filled=True, rounded=True)
plt.title("Decision Tree for House Price Prediction")
plt.show()
```
